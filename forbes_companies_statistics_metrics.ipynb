```python
import pandas as pd
import pycountry_convert as pc
import numpy as np
import statistics as sts
```

```python
dataset=pd.read_csv('Forbes_2000_top_company_CLNQ11.csv',delimiter=',')
dataset.head()
```

```python
dataset1=dataset.copy()
dataset1
```

```python
dataset1=dataset1[dataset1['Year Founded']!=0]
dataset1
```

```python
dataset1=dataset1[dataset1['Total Employees']!=0]
dataset1
```

```python
country=[dataset1['Country']]
country[0]
```

```python
def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name
```

```python
continents=[]
for i in country[0]:
    continents.append(country_to_continent(i))
continents
```

```python
dataset1['Continents']=continents
```

```python
dataset1['Continents'].value_counts(normalize=True)*100
```

```python
profits=[round(dataset1[dataset1['Continents']=='Asia']['Profits (Billions)'].sum(),2),
round(dataset1[dataset1['Continents']=='North America']['Profits (Billions)'].sum(),2),
round(dataset1[dataset1['Continents']=='Europe']['Profits (Billions)'].sum(),2),
round(dataset1[dataset1['Continents']=='South America']['Profits (Billions)'].sum(),2),
round(dataset1[dataset1['Continents']=='Oceania']['Profits (Billions)'].sum(),2),
round(dataset1[dataset1['Continents']=='Africa']['Profits (Billions)'].sum(),2)]
```
